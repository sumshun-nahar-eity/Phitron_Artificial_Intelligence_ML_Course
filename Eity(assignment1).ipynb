```python
import math

class Shape:

    def area(calc):

        return 0

class Rectangle(Shape):

    def __init__(calc, w, h):
        calc.w = w
        calc.h = h

    def area(calc):
        return calc.w * calc.h

class Circle(Shape):

    def __init__(calc, r):
        calc.r = r

    def area(calc):

        return math.pi * calc.r * calc.r


data = input().split()


if data[0] == "Rectangle":

    w = float(data[1])

    h = float(data[2])

    obj = Rectangle(w, h)

    print(f"Area: {obj.area():.2f}")

if data[0]=="rectangle":

    print("Please write Rectangle for the output")

if data[0]=="circle":

    print("Please write Circle for the output")

elif data[0] == "Circle":

    r = float(data[1])

    obj = Circle(r)

    print(f"Area: {obj.area():.2f}")


```

```python
import math

number_taken = int(input("Input:"))

summation = 0

for d in str(number_taken):

    summation = summation + math.factorial(int(d))


if summation == number_taken:

    print("Output: Strong Number")

else:
    print("Output: Not Strong Number")
```
