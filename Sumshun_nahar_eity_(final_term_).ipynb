## Question : 1
20 Marks
Load the TrainingHours column into a NumPy array.
Convert to float datatype


Find mean & standard deviation

Final output: Two numeric values






```python
import numpy as np
import pandas as pd

url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/main/final-employee-ds.csv"
df = pd.read_csv(url)

training_hours = df["TrainingHours"].astype(float).to_numpy()
mean = np.mean(training_hours)
std_value = np.std(training_hours)
print("Mean:", mean)
print("Standard Deviation:", std_value)
```

## Question : 2
20 Marks
Handle missing values (if any):
Fill missing Salary with department-wise median


Calculate total Salary sum for employees with ExperienceYears > 12 and ProjectCount > 5

Final output: One numeric value t





```python
import pandas as pd

url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/main/final-employee-ds.csv"
df = pd.read_csv(url)

df["Salary"] = df["Salary"].fillna(df.groupby("Department")["Salary"].transform("median"))

filtered_salary = df.loc[(df["ExperienceYears"] > 12) & (df["ProjectCount"] > 5), "Salary"]

print(filtered_salary.sum())
```

## Question : 3
20 Marks

Filter employees who:
Department = "IT" or "Finance"


Age between 30–45


PerformanceScore > 88


Sort by Salary descending

Final output: Table with Name, Department, Salary, PerformanceScore





```python
import pandas as pd

url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/main/final-employee-ds.csv"
df = pd.read_csv(url)

filtered = df[
    (df["Department"].isin(["IT", "Finance"])) &
    (df["Age"] >= 30) & (df["Age"] <= 45) &
    (df["PerformanceScore"] > 88)
]

output = filtered.sort_values("Salary", ascending=False)[["Name", "Department", "Salary", "PerformanceScore"]]
print(output.to_string(index=False))
```

## Question : 4
20 Marks
Add new column SalaryPerHour = Salary ÷ (WorkHoursPerWeek × 4)
Filter top 5 employees by SalaryPerHour

Final output: Table(5 Rows) with Name, SalaryPerHour







```python
import pandas as pd

url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/main/final-employee-ds.csv"
df = pd.read_csv(url)

df["SalaryPerHour"] = df["Salary"] / (df["WorkHoursPerWeek"] * 4)

output = df.nlargest(5, "SalaryPerHour")[["Name", "SalaryPerHour"]]
print(output.to_string(index=False))
```

## Question : 5
20 Marks
Create a bar chart showing the count of employees in each Department, separated by Gender. You can solve the question with any of the package (Matplotlib , Seaborn , Plotly )
Dataset Columns Used:
Department → Categorical variable (IT, Finance, HR, Marketing, Management)


Gender → Categorical variable (Male, Female)

Expected Output
A bar chart where:


x-axis → Department


y-axis → Number of employees


Each bar split into Male and Female segments


Colors differentiate gender






```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/main/final-employee-ds.csv"
df = pd.read_csv(url)

plt.figure(figsize=(12, 7))
sns.countplot(x="Department", hue="Gender", data=df)
plt.title("Count of Employees by Dept & Gender")
plt.xlabel("Department")
plt.ylabel("Employees")
plt.xticks(rotation=30)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```
